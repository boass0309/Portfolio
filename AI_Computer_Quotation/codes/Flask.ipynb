```python
from flask import Flask
import threading
# Thred 생성으로 Flask 애플리케이션이 백그라운드에서 실행됨

app = Flask(__name__)
@app.route('/create')
def create():
    return 'Created'

def template(contents, content):
    print(content)
    return f'''
        <html>
            <head>
                <title>AI_Qutation</title>
            </head>
            <body>
                <h1>Data Table</h1>
                <ol>
                    {contents}
                </ol>
                {content}
            </body>
        </html>
        '''
def getContents():
    liTags = ''
    for topic in topics:
        liTags = liTags + f'<li><a href="/read/{topic["id"]}/">{topic["title"]}</a></li>'
    return liTags

@app.route('/read/<int:id>/') # id가 함수의 파라미터로 전달됨
def read(id):
    title = ''
    body = ''  
    for topic in topics:
        if id == topic['id']:
            title = topic['title']
            body = topic['body']
            break
    return template(getContents(),f'<h2>{title}</h2>{body}')
    

topics = [
    {'id': 1, 'title':'html', 'body': 'html is ...'},
    {'id': 2, 'title':'css', 'body': 'css is ...'},
    {'id': 3, 'title':'javascript', 'body': 'javascript is ...'}
]

@app.route('/')
def index():
    intel = pd.read_csv('intel.csv', encoding='cp949')
    amd = pd.read_csv('amd.csv', encoding='cp949')
    # 따옴표 3개는 여러 줄에 걸친 ' ' 이라고 보면 됨
    return template(getContents(), f'<h2>Welcome</h2>Hello, WEB1')
    


def run_app():
    app.run(port=5000, debug=True, use_reloader=False)
    # use_reloader: 코드 변경 시 서버가 두 번 실행되는 옵션

if __name__ == "__main__":
    thread = threading.Thread(target=run_app)
    thread.start()
```
